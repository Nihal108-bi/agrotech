Key Columns:
Temperature (Celsius and Fahrenheit): Current temperature readings.
Wind Speed (mph and kph): Measures wind speed.
Pressure (mb and in): Atmospheric pressure readings.
Precipitation (mm and in): Indicates the amount of rainfall.
Humidity (%): Shows the humidity level.
Cloud (%): Represents cloud cover.
Visibility (km and miles): Distance one can see in clear weather.
UV Index: Measures the intensity of ultraviolet rays.
Gust (mph and kph): Wind gusts' speed.

```python
## segregate numerical and categorical columns

numerical_columns=df.columns[df.dtypes!='object']
categorical_columns=df.columns[df.dtypes=='object']
print("Numerical columns:",numerical_columns)
print('Categorical Columns:',categorical_columns)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
x=0
for i in numerical_columns:
    sns.histplot(data=df,x=i,kde=True)
    print('\n')
    plt.show()
```
